```python
#import needed packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
from scipy import stats

#install and show data 
#(as we need only two columns, I made new data frame from copying two columns from the original dataset)

df = pd.read_csv('kc_house_data.csv')
new = df[["yr_built", "sqft_living"]].copy()
print(new.head(5))
```

```python
#show statistics of the new data set which consists of two necessary columns
print(new.describe()) #use pandas method .describe() to show some statistics of the data

def morestatistics(column1): #create a function to show other statistics of the data
    column = list(column1.values) #make list from the data frame values
    mode = max(column, key=column.count) #search for the values with the max number of occurance
    print('mode = ' + str(mode)) #show mode
    column.sort() #put values in order from min to max
    med1=column[len(column)//2] #find the first middle value through integer division
    med2=column[len(column)//2-1] #find the second middle value (as we know that the dataset has odd â„– of rows)
    median_column=(med1+med2)/2 #calculate median as the arithmetical mean of 2 middle values
    print('median = ' + str(median_column)) #show median

    #perform our function to two columns in our data frame
print('yr_built other statistics:')
morestatistics(new["yr_built"]) 
print('sqft_living other statistics:')
morestatistics(new["sqft_living"])
```

```python
#visualization
year_built = list(new["yr_built"]) #make two lists from the values of two columns
area = list(new["sqft_living"])

#make two independent figures with the histograms of two created lists
fig1 = plt.figure()  #make 1st figure
ax1 = fig1.add_subplot(1, 1, 1) #add subplot on the figure with the coordinates (1,1,1) - default
nyam = ax1.hist(year_built, bins=40) #make histograms for the years with 40 bins for accuracy
ax1.set_xlabel('Year when the house was built (year)') #label x-axis
ax1.set_ylabel('Frequency') #label y-axis

fig2 = plt.figure()  #make 2nd figure
ax2 = fig2.add_subplot(1, 1, 1)  figure
ax1 = fig1.add_subplot(1, 1, 1) #add subplot on the figure for living areas with 40 bins for accuracy
ax2.set_xlabel('Living area (square feet)') #label x-axis
ax2.set_ylabel('Frequency') #label y-axis
```

```python
#confidence interval
def confidence_interval(sample_mean, SD, sample_size): #make function for confidence interval
    st_error = SD/np.sqrt(sample_size)
    print ('SE = '+ str(st_error))
    #determine Standard Error using SD and n
    min_mean = sample_mean - (1.96*st_error) #calculate the minimum value in confidence interval
    min_mean = round(min_mean, 2) #round this value
    max_mean = sample_mean + (1.96*st_error) #calculate the maximum value in confidence interval
    max_mean = round(max_mean, 2) #round this value
    print('95% confidence interval is ('+ str(min_mean) + ', ' + str(max_mean) + ')') #show confidence interval
    
#use function with the values calculated earlier
confidence_interval(2079.9, 918.44, 21613) 
```

```python
#subgroups examination
new_house = [] #create 2 empty lists to filter values for old and young houses
old_house = []

for i in range(21613): #make the loop to filter the values of living area based on the age
    if year_built[i] > 1989:
        new_house.append(area[i])
    else:
        old_house.append(area[i])
#print some of the statistics for old and young houses separately using functions from statistics package and other
print('The number of houses younger than 30 years old is ' + str(len(new_house))) 
print('The mean of young houses is ' + str(round(statistics.mean(new_house), 3)))
print('The standart deviation of young houses is ' + str(round(statistics.stdev(new_house), 3)))
print('The number of houses older or equal to 30 years old is ' + str(len(old_house)))
print('The mean of old houses is '+ str(round(statistics.mean(old_house), 3)))
print('The standart deviation of old houses is ' + str(round(statistics.stdev(old_house), 3)))

#create two separate figures with histograms for young and old houses
fig1 = plt.figure()
ax1 = fig1.add_subplot(1, 1, 1)
nyam = ax1.hist(new_house, bins=40)
ax1.set_xlabel('Living area for young houses (square feet)')
ax1.set_ylabel('Frequency')

fig2 = plt.figure()
ax2 = fig2.add_subplot(1, 1, 1)
nyam = ax2.hist(old_house, bins=40)
ax2.set_xlabel('Living area for old houses (square feet)')
ax2.set_ylabel('Frequency')
```

```python
#difference of means test
def difference_of_means(list1,list2,number_of_tails): #create function for t-test for difference of means
    m1=statistics.mean(list1) #use some functions from statistics package to calculate necessary values
    sd1=statistics.stdev(list1)
    n1=len(list1)
    m2=statistics.mean(list2)
    sd2=statistics.stdev(list2)
    n2=len(list2)
    SE=np.sqrt(((sd1**2)/n1)+((sd2**2)/n2)) #calculate Standard Error for the difference of mean
    print('Standart error for the difference of means is ' + str(round(SE, 3))) #print rounded standard error
    T=((m1-m2)-0)/SE # calculate t-score of the difference of mean
    print('t-score for the difference of means is ' + str(round(T, 3))) #print rounded t-score
    if n1 > n2: #calculate degrees of freedom with the loop
        doff = n2-1
    else:
        doff = n1-1
    print('Degrees of freedom is ' + str(doff)) #print degrees of freedom
    p = (1 - stats.t.cdf(T, df=doff))*number_of_tails #calculate p-value with scipy.stats function
    print('p-value for the difference of means is ' + str(p)) #print p-value
    s = np.sqrt((((n1 - 1) * (sd1**2)) + ((n2 - 1) * (sd2**2))) / (n1 + n2 - 2)) #calculate pooled and weighted SD
    hg=(m1-m2)/s #calculate hedge'd g (effect size)
    print('Effect size (hedge\'s g) is ' +str(round(hg, 6))) #print hedge's g
    
difference_of_means(new_house, old_house, 2) #use function for the old and young houses
```
